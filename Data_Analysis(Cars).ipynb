1). Instruction(For Data Cleaning)

 *   Find all Null Value in the datset  if there is any null value in any column, then fill it with the mean of that column.
 









```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(),inplace=True) # inplace is used for permanent Changes.
```
