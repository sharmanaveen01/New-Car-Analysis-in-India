```python
new_car['Rating(outof10)'].replace('na',5,inplace=True)
```

```python
new_car['Rating(outof10)'] = new_car['Rating(outof10)'].astype(float)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = 'Style' , y = 'Rating(outof10)',data = new_car, palette = 'hls')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Transmission', y = 'Rating(outof10)', data=new_car,palette='hls')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='VehicleType', y ='Rating(outof10)', data = new_car, palette='hls')
plt.xticks(rotation=90)
plt.show()
```

```python
new_car_data = new_car.sort_values(by=['Rating(outof10)'], ascending=False)
```

```python
new_car_data = new_car_data.reset_index()
```

```python
new_car_data = new_car_data.drop(['index'], axis =1)
```

```python
new_car_data.head()
```

```python
new_car_data.tail()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='Style', y='Rating(outof10)', data = new_car_data,palette='hls')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x = 'Transmission', y ='Rating(outof10)',data = new_car_data,palette='hls')
plt.xticks(rotation=90)
plt.show()
```
